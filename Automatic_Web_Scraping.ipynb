```python
from autoscraper import AutoScraper

url = 'https://github.com/RishavMishraRM?tab=repositories'

wanted_list = ['GRADIO_Inception_Net', 'Python']

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar('https://github.com/RishavMishraRM?tab=repositories', grouped=True)
```

```python
scraper.set_rule_aliases({'rule_dfkd' : 'Title', 'rule_45w1' : 'Language Used'})
scraper.keep_rules(['rule_dfkd', 'rule_45w1'])
scraper.save('github-repository-search')
```

```python
scraper.load('github-repository-search')
```
